```python
main_data["Engagement_Score"] = main_data["Swipes"] + main_data["Messages_Sent"]
main_data.head()
```

```python
average_engagement = main_data['Engagement_Score'].mean()
max_engagement = main_data['Engagement_Score'].max()
min_engagement = main_data['Engagement_Score'].min()

print(f'Average Engagement trend is: {average_engagement}')
print(f'Maximum Engagement trend is: {max_engagement}')
print(f'Minimum Engagement trend is: {min_engagement}')
```

```python
# Calculate standard deviation and variance
std_dev_engagement = main_data["Engagement_Score"].std()
variance_engagement = main_data["Engagement_Score"].var()

# Print results
print(f"Standard Deviation of Engagement Scores: {std_dev_engagement}")
print(f"Variance of Engagement Scores: {variance_engagement}")
```

```python
# Sort the DataFrame by Engagement_Score in descending order
main_data_sorted = main_data.sort_values(by="Engagement_Score", ascending=False)

main_data_sorted
```

```python
# Calculate the 75th percentile of the Engagement Scores
percentile_75 = main_data["Engagement_Score"].quantile(0.75)

# Filter users with Engagement Score greater than or equal to the 75th percentile
top_25_percent_users = main_data[main_data["Engagement_Score"] >= percentile_75]

# Display the top 25% users
top_25_percent_users.reset_index()
```

```python
Age_18_25 = main_data[(main_data['Age']>=18) & (main_data['Age']<=25)]
Age_26_35 = main_data[(main_data['Age']>=26) & (main_data['Age']<=35)]

Age_18_25.mean()
```

```python
# Divide users into age groups
main_data['Age_Group'] = pd.cut(main_data['Age'], bins=[18, 25, 35], labels=['18-25', '26-35'])

# Calculate the average engagement score for each age group
average_engagement_by_group = main_data.groupby('Age_Group')['Engagement_Score'].mean()

# Display the results
print(average_engagement_by_group)
```

```python
main_data['Match_Rate'] = ((main_data['Matches']/main_data['Swipes'])*100).round(2)
```

```python
main_data.head()
```

```python
# Calculate the correlation between Match Rate and Engagement Score
correlation = main_data['Match_Rate'].corr(main_data['Engagement_Score'])

# Print the correlation value
print(f"Correlation between Match Rate and Engagement Score: {correlation}")
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot to visualize the relationship
plt.scatter(main_data['Engagement_Score'], main_data['Match_Rate'])
plt.title('Engagement Score vs Match Rate')
plt.xlabel('Engagement Score')
plt.ylabel('Match Rate (%)')
plt.show()
```

```python
# Create quartiles based on Engagement Score
main_data['Engagement_Quartile'] = pd.qcut(main_data['Engagement_Score'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

# Calculate the average Match Rate for each quartile
avg_match_rate_by_quartile = main_data.groupby('Engagement_Quartile')['Match_Rate'].mean()

# Display the results
print(avg_match_rate_by_quartile)
```
